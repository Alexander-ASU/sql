# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
rng = np.random.RandomState(123)

coin_flips = rng.randint(0, 6, size=10)
ax= sns.distplot(coin_flips,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='y', ylabel='Вероятность')
plt.title('Плотность рапределения подбрасывания кубика')
plt.show()
heads = np.sum(coin_flips)
#print('Количество выпавшей решки: ',heads)
```

Логнормальное распределение

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
rng = np.random.RandomState(100)

coin_flips = rng.randint(0, 6, size=1000)
ax= sns.distplot(coin_flips,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='y', ylabel='Вероятность')
plt.title('Плотность рапределения подбрасывания кубика')
plt.show()
heads = np.sum(coin_flips)
#print('Количество выпавшей решки: ',heads)
```

График изменился так как при увеличении попыток распределение стремиться к нормальному.
