### Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
mu = 2.0
sigma = 0.5

mu1 = 3.0
sigma1 = 1
# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv1 = sts.norm(loc=mu1, scale=sigma1)

x = np.linspace(0,4,100)
x1 = np.linspace(0,5,100)
x2 = np.linspace(0,3,100)

pdf = norm_rv.pdf(x)
pdf1 = norm_rv1.pdf(x1)

plt.plot(x, pdf, x1, pdf, x1, pdf1, x2, pdf1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
lam = 0.1
lam1 = 0.5
lam2 = 1.0
expon_rv = sts.expon(lam)
expon_rv1 = sts.expon(lam1)
expon_rv2 = sts.expon(lam2)
expon_rv.rvs(10)
```

```python
x = np.linspace(0,5,1000)
x1 = np.linspace(0,3,1000)
x2 = np.linspace(0,1,1000)
pdf = expon_rv.pdf(x)
pdf1 = expon_rv1.pdf(x)
pdf2 = expon_rv2.pdf(x)
plt.plot(x, pdf, x1, pdf1, x2, pdf1, x1, pdf2, x2, pdf2)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
n = 10
n1 = 2

t_rv = sts.t(n)
t_rv1 = sts.t(n1)

t_rv.rvs(10)
```

```python
x = np.linspace(-5,5,1000)
x1 = np.linspace(-5,3,1000)
pdf = t_rv.pdf(x)
pdf1 = t_rv1.pdf(x)

plt.plot(x, pdf, x1, pdf, x, pdf1, x1, pdf1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
