# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd
data = pd.read_csv('power.csv')
#data
#data['country'].unique()
```

```python
rez = data[(data['quantity'] >= 0)]
```

```python
rez = rez[(rez['year'] <= 2010) & (rez['year'] >= 2005)]
```

```python
rez = rez[(rez['category'] == 4) | (rez['category'] == 12) | (rez['category'] == 21) ]
```

```python
rez = rez[(rez['country'] == 'Latvia') | (rez['country'] == 'Lithuania') | (rez['country'] == 'Estonia')]
rez
```

```python
rez['quantity'].sum()
```
